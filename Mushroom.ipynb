# Seperating X(Feature) and Y (Target)

```python
features=DF.drop("class",axis=1)
target=DF["class"]
```

```python
from sklearn.linear_model import LogisticRegression
Log=LogisticRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
for i in range(0,100):
    X_train,X_test,Y_train,Y_test=train_test_split(features,target,test_size=0.2,random_state=i)
    Log.fit(X_train,Y_train)
    pred_train=Log.predict(X_train)
    pred_test=Log.predict(X_test)
    print(f"At random state {i},the training accuracy is:- {r2_score(Y_train,pred_train)}")
    print(f"At random state {i},the testing accuracy is:- {r2_score(Y_test,pred_test)}")
    print("\n")
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,target,test_size=0.2,random_state=28)
```

```python
Log.fit(X_train,Y_train)
```

```python
pred_test=Log.predict(X_test)
```

```python
print(r2_score(Y_test,pred_test))
```

```python
Accuracy_Train=r2_score(Y_train,pred_train)
Accuracy_Test=r2_score(Y_test,pred_test)

from sklearn.model_selection import cross_val_score
for j in range(2,10):
    cvscore=cross_val_score(Log,features,target,cv=j)
    cvmean=cvscore.mean()
    print(f"At cross fold {j} the cv score is {cvmean} and accuracy score for training is {Accuracy_Train} and accuracy for the testing is {Accuracy_Test}")
    print("\n")
```

```python
#plt.figure(figsize=(8,6))
plt.scatter(x=Y_test, y=pred_test, color='r')
plt.plot(Y_test,Y_test, color='b')
```

```python
from sklearn.linear_model import LogisticRegression

## lr = LogisticRegression(solver="lbfgs")
Log = LogisticRegression(solver="liblinear")
Log.fit(X_train,Y_train)

print("Test Accuracy: {}%".format(round(Log.score(X_test,Y_test)*100,2)))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

best_Kvalue = 0
best_score = 0

for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    if knn.score(X_test,Y_test) > best_score:
        best_score = knn.score(X_train,Y_train)
        best_Kvalue = i

print("""Best KNN Value: {}
Test Accuracy: {}%""".format(best_Kvalue, round(best_score*100,2)))

rfscore = cross_val_score(knn,features,target,cv=5)
rfc = rfscore.mean()
print('Cross Val Score:',rfc*100)
```

# SVC

```python
from sklearn.svm import SVC

svm = SVC(random_state=28, gamma="auto")
svm.fit(X_train,Y_train)

print("Test Accuracy: {}%".format(round(svm.score(X_test,Y_test)*100,2)))

rfscore = cross_val_score(svm,features,target,cv=5)
rfc = rfscore.mean()
print('Cross Val Score:',rfc*100)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=100, random_state=28)
RF.fit(X_train,Y_train)

print("Test Accuracy: {}%".format(round(RF.score(X_test,Y_test)*100,2)))

rfscore = cross_val_score(RF,features,target,cv=5)
rfc = rfscore.mean()
print('Cross Val Score:',rfc*100)
```
